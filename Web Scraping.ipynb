# Web Scraping

# Accessing the HTML content from webpage

```python
import requests 
URL = "https://www.zinavo.com/"
r = requests.get(URL) 
print(r.content) 
```

# Parsing the HTML content

```python
import requests 
from bs4 import BeautifulSoup 
  
URL = "https://www.zinavo.com/"
r = requests.get(URL) 
  
soup = BeautifulSoup(r.content, 'html5lib') 
print(soup.prettify()) 
```
